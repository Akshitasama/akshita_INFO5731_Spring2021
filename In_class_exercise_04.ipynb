**1.1 Feature Extraction**

```python
# Write your code here
from six.moves import urllib
import pandas as pd
a = []
data_link = "https://raw.githubusercontent.com/unt-iialab/info5731_spring2021/main/class_exercises/01-05-1%20%20Adams%20v%20Tanner.txt"
data_files = urllib.request.urlopen(data_link)

for row in data_files:
  changed_data = row.decode("utf-8").replace('\r\n', '')
  if changed_data: a.append(changed_data)
data_frame = pd.DataFrame (a, columns = ['Raw Data'])
print(data_frame)

```

**Number of Sentences**

```python
data_frame['Number of Sentences'] = data_frame['Raw Data'].apply(lambda x: len(str(x).split(".")))
data_frame
```

**Number of Words**

```python
data_frame['Number of Words']=data_frame['Raw Data'].apply(lambda x: len(str(x).split(" ")))
data_frame
```

**Number of Characters**

```python
data_frame['Number of Characters']=data_frame['Raw Data'].str.len()
data_frame
```

**Average Word Length**

```python
def avg_word(sentence):
  w=sentence.split()
  if len(w)!=0:
    return (sum(len(word) for word in w)/len(w))
  else:
    return None

data_frame['Average word length']=data_frame['Raw Data'].apply(lambda x: avg_word(x))
data_frame
```

**Number of Stopwords**

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
s=stopwords.words('english')
data_frame['NUmber of stopwords']= data_frame['Raw Data'].apply(lambda x: len([x for x in x.split() if x in s]))
data_frame
```

**Number of special Characters**

```python
def spcl_char(sentence):
  a = 0
  for x in range(len(sentence)):
    if not((sentence[x].isalpha()) and (sentence[x].isdigit())):
        a = a + 1
  return a

data_frame['Number of special characters'] = data_frame['Raw Data'].apply(lambda x: spcl_char(x))
data_frame
```

**Number of numerics**

```python
data_frame['Numbers of numerics'] = data_frame['Raw Data'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
data_frame
```

**Number of Upper case**

```python
data_frame['Number of upper case words'] = data_frame['Raw Data'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
data_frame
```

**1.2 Basic Text Pre-processing of text data**

```python
data_frame['Lower Case'] = data_frame['Raw Data'].apply(lambda x: " ".join(x.lower() for x in x.split()))
data_frame
```

**Punctuation Removal**

```python
data_frame['Punctuation Removal'] = data_frame['Lower Case'].str.replace('[^\w\s]','')
data_frame
```

**Stopwords Removal**

```python
from nltk.corpus import stopwords
s = stopwords.words('english')
data_frame['Removal of Stopwords'] = data_frame['Punctuation Removal'].apply(lambda x: " ".join(x for x in x.split() if x not in s))
data_frame
```

**Frequent words removal**

```python
frequent_words = pd.Series(' '.join(data_frame['Removal of Stopwords']).split()).value_counts()[:10]
print(frequent_words)
freq_words_list = list(frequent_words.index)
data_frame['Common words removed'] = data_frame['Removal of Stopwords'].apply(lambda x: " ".join(x for x in x.split() if x not in freq_words_list))
data_frame
```

**Rare words removal**

```python
rare= pd.Series(' '.join(data_frame['Removal of Stopwords']).split()).value_counts()[-10:]
print(rare)
rare_list = list(rare.index)
data_frame['Rare Words removed'] = data_frame['Common words removed'].apply(lambda x: " ".join(x for x in x.split() if x not in rare_list))
data_frame
```

**Spelling Correction**

```python
from textblob import TextBlob
data_frame['Correction of Spelling'] = data_frame['Rare Words removed'].apply(lambda x: str(TextBlob(x).correct()))
data_frame
```

**Tokenization**

```python
import nltk
nltk.download('punkt')
data_frame['Tokenization'] = data_frame['Correction of Spelling'].apply(lambda x: TextBlob(x).words)
data_frame
```

**Stemming**

```python
from nltk.stem import PorterStemmer
s = PorterStemmer()
data_frame['After Stemming'] = data_frame['Tokenization'].apply(lambda x: " ".join([s.stem(word) for word in x]))
data_frame
```

**Lemmatization**

```python
from textblob import Word
import nltk
nltk.download('wordnet')

data_frame['After Lemmatization'] = data_frame['After Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
data_frame
```

**1.3 Saving to CSV**

```python
data_frame.to_csv('Final_cleaned_data.csv',index=False)
```

**1.4 Advanced Text Processing**

**Calculating Term frequencies**

```python
text=(data_frame['After Lemmatization']).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
text.columns=['words', 'tf']
text
```

**Top 10- one grams**

```python
import itertools
from nltk import ngrams
import collections
from nltk.tokenize import word_tokenize

w_sen= []
for row in data_frame['After Lemmatization']:
  w_sen.append(word_tokenize(row))
cleaningdone = [x for x in w_sen if x != []]
i = list(itertools.chain.from_iterable(cleaningdone))
one = ngrams(i, 1)
collections.Counter(one).most_common(10)
```

**Top 10- Bi grams**

```python
bi_gram = nltk.bigrams(i)
f_dist = nltk.FreqDist(bi_gram)
(f_dist.most_common(10))
```

**Top 10- Tri grams**

```python
t_grams=nltk.trigrams(i)
f_dist=nltk.FreqDist(t_grams)
f_dist.most_common(10)
```
