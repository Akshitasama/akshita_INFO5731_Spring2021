```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
main_heading=[] #creating list to store all the names of the reviews.
body=[] #List to store the description of the reviews.
rate=[]# To store the rating of the products
for number in range(50):
  html_page=requests.get("https://www.flipkart.com/realme-narzo-20-glory-silver-64-gb/product-reviews/itm4ac58d879006d?pid=MOBFVEATBBRGJBKH&lid=LSTMOBFVEATBBRGJBKHO9FHOE&marketplace=FLIPKART").text #generating link and accessing webpage
  soup=BeautifulSoup(html_page,'lxml') #creating instance for the BeautifulSoup library
  review_name=soup.find_all(class_='_2-N8zT') #getting name of the reviews by using class name
  review_desc=soup.find_all(class_='t-ZTKy') #getting description of the reviews by class name
  rating=soup.find_all(class_='_3LWZlK _1BLPMq') #getting ratings by using class name
  for a, b, c in zip(review_name, review_desc,rating): #iterating the list
    main_heading.append(a.text) #appending the required elements to empty list
    body.append(b.text)
    rate.append(c.text)
data_fr=pd.DataFrame(list(zip(main_heading, body, rate)), columns=['Review Title','Review description', 'Rating']) #dataframe creation
print("Length of data frame is {0}".format(len(data_fr))) #printing lenth
data_fr
#data_fr.to_csv('data.csv', index=False) # Storing in the form of csv file from dataframe
```

# **Preprocessing Data**

**Converting to lower case**

```python
data_fr['After Preprocessing'] = data_fr['Review description'].apply(lambda x: " ".join(x.lower() for x in x.split()))
data_fr
```

**Removing Punctuation**

```python
data_fr['After Preprocessing'] = data_fr['After Preprocessing'].str.replace('[^\w\s]','')
data_fr
```

**Removing Numerics**

```python
data_fr['After Preprocessing'] = data_fr['After Preprocessing'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
```

**Removing Special Characters**

```python
import re
data_fr['After Preprocessing'] = data_fr['After Preprocessing'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))
data_fr
```

```python
import nltk
nltk.download()
```

**Removing Stop Words**

```python
from nltk.corpus import stopwords
s = stopwords.words('english')
data_fr['After Preprocessing'] = data_fr['After Preprocessing'].apply(lambda x: " ".join(x for x in x.split() if x not in s))
data_fr
```

**Stemming**

```python
from nltk.stem import PorterStemmer
stem_ = PorterStemmer()
data_fr['After Preprocessing'].apply(lambda x: " ".join([stem_.stem(word) for word in x.split()]))
```

**Lemmetization**

```python
from textblob import Word
import nltk
nltk.download('wordnet')
data_fr['After Preprocessing'] = data_fr['After Preprocessing'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
data_fr
```

# **Parts of Speech Tagging and Features**

```python
from nltk.tokenize import word_tokenize
parts_speech= []
for row in data_fr['After Preprocessing']:
  t = word_tokenize(row)
  parts_speech.append(nltk.pos_tag(t))
parts_speech
```

```python
Adjective_ = []
Adverb_ = []
Cordinating_Conjunction = []
Sub_ordinating_Conjuction = []
Interjection_ = []
Noun_ = []
Verb_ = []
PersonalPronoun_ = []
predeterminer_ = []
Determiner_= []
```

```python
for value in parts_speech:
  Adj_Count = 0
  Adv_Count = 0
  CC_Count = 0
  SC_Count = 0
  I_Count = 0
  Noun_Count = 0
  Verb_Count = 0
  PP_Count = 0
  predeterminer_Count = 0
  Determiner_Count = 0
  for word,tag in value:
    if tag == 'JJ':
      Adj_Count= Adj_Count + 1
    elif tag == 'RB':
      Adv_Count = Adv_Count + 1
    elif tag == 'CC':
      CC_Count = CC_Count + 1
    elif tag == 'UH':
      I_Count = I_Count + 1
    elif tag == 'NN':
      Noun_Count = Noun_Count + 1
    elif tag == 'VR':
      Verb_Count = Verb_Count + 1
    elif tag == 'PRP':
      PP_Count = PP_Count + 1
    elif tag == 'PDT':
      predeterminer_Count = predeterminer_Count + 1
    elif tag == 'DT':
      Determiner_Count = Determiner_Count + 1
    elif tag == 'IN':
      SC_Count = SC_Count + 1
  Adjective_.append(Adj_Count)
  Adverb_.append(Adv_Count)
  Cordinating_Conjunction.append(CC_Count)
  Interjection_.append(I_Count)
  Noun_.append(Noun_Count)
  Verb_.append(Verb_Count)
  PersonalPronoun_.append(PP_Count)
  predeterminer_.append(predeterminer_Count)
  Determiner_.append(Determiner_Count)
  Sub_ordinating_Conjuction.append(SC_Count)
```

```python
data_fr['Number of Adjectives'] = Adjective_
data_fr['Number of Adverbs'] = Adverb_
data_fr['Number of Cordinating Conjunctions'] = Cordinating_Conjunction
data_fr['Number of Interjections'] = Interjection_
data_fr['Number of Nouns'] = Noun_
data_fr['Number of Verbs'] = Verb_
data_fr['Number of Personal Pronouns'] = PersonalPronoun_
data_fr['Number of Predeterminers'] = predeterminer_
data_fr['Number of Determiners'] = Determiner_
data_fr['Number of Subordinating Conjuctions'] = Sub_ordinating_Conjuction
data_fr
```
